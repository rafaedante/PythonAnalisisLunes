```python
# graficar un polinomio de grado2 (y = a(x**2) + bx+ c)
x = np.linspace(-5, 15)
a = 5
b = 2
c = 15
y = a*(x**2) + b*x+ c

plt.plot(x, y)
plt.grid(True)
plt.title('y = ax + b')
plt.xlabel('eje x')
plt.ylabel('eje y')
```
