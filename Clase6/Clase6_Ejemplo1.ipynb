```python
# graficar un polinomio de grado2 (y = a(x**2) + bx+ c)
x = np.linspace(-5, 15)
a = 5
b = 2
c = 15
y = a*(x**2) + b*x+ c

plt.plot(x, y)
plt.grid(True)
plt.title('y = a*(x**2) + b*x+ c')
plt.xlabel('eje x')
plt.ylabel('eje y')
```

```python
x = np.linspace(0, np.pi*4, 100)
#x
seno = np.sin(x)
coseno = np.cos(x)

plt.plot(x, seno)
plt.plot(x, coseno)
plt.grid(True)
```

```python
x = np.linspace(0,10,10)
plt.plot(np.concatenate((-x,x)), np.concatenate(((-x)**2, x**2)), 'mD--')
plt.plot(np.concatenate((-x,x)), np.concatenate(((-x)**3, x**3)), 'cd-')
```

```python
a = np.linspace(0,20,50)
b= np.sin(a)
plt.figure()

#grafico1
plt.subplot(2,2,1)
plt.plot(a,b*25,'r')

#grafico2
plt.subplot(2,2,2)
plt.plot(a+2,b,'g')

#grafico3
plt.subplot(2,2,3)
plt.plot(a,b,'b')

#grafico4
plt.subplot(2,2,4)
plt.plot(b,a,'y')

plt.grid(True)
plt.show()
```
